## Problem 1

Let $f(x) = 2\sin x - \cos{2x}$. Write a function `beauty_derivative_plot` that takes a point $x$ and plots the function $f(x)$ and its derivative $f'(x)$ in the range $x \in [-2\pi, 2\pi]$. The function should also plot the tangent line to the function at the point $x$.

Example of output for $x = -3, -2, 1.7$:

```python
beauty_derivative_plot(-3)
```
<img src="https://i.ibb.co/W4pyPfTX/tangent-line-3-0.png" width="600"/>

```python
beauty_derivative_plot(-2)
```
<img src="https://i.ibb.co/F4zDq8m0/tangent-line-2-0.png" width="600"/>

```python
beauty_derivative_plot(1.7)
```
<img src="https://i.ibb.co/TBmLXg2h/tangent-line-1-7.png" width="600"/>

````python
import matplotlib.pyplot as plt
import numpy as np
def beauty_derivative_plot(x_point):
    # Определяем функцию f(x)
    def f(x):
        return 2 * np.sin(x) - np.cos(2 * x)

    # Определяем производную f'(x)
    def f_prime(x):
        return 2 * np.cos(x) + 2 * np.sin(2 * x)

    # Определяем уравнение касательной в точке x_point
    def tangent_line(x):
        return f(x_point) + f_prime(x_point) * (x - x_point)

    # Генерируем значения x для построения графиков
    x_values = np.linspace(-2 * np.pi, 2 * np.pi, 1000)

    # Вычисляем значения y для f(x), f'(x) и касательной
    y_f = f(x_values)
    y_f_prime = f_prime(x_values)
    y_tangent = tangent_line(x_values)

    # Строим график функции f(x)
    plt.plot(x_values, y_f, label=r'$f(x) = 2\sin x - \cos 2x$',color='blue', linewidth = 1)

    # Строим график производной f'(x)
    plt.plot(x_values, y_f_prime, label=r"$f'(x) = 2\cos x + 2\sin 2x$", color = 'red', linewidth = 1, linestyle='--')

    # Строим касательную в точке x_point
    tangent_x_values = np.linspace(x_point - 1, x_point + 1, 100)
    y_tangent = tangent_line(tangent_x_values)
    plt.plot(tangent_x_values, y_tangent, color='green',label=f'Tangen at $x = {x_point}$', linewidth = 1)

    # Добавляем пунктирные линии от точки касания до осей
    plt.plot([x_point, x_point], [0, f(x_point)], color='purple', linestyle='--', linewidth=0.8)
    plt.plot([0, x_point], [f(x_point), f(x_point)], color='purple', linestyle='--', linewidth=0.8)
    # Добавляем подпись с координатами точки касания и стрелку
    plt.annotate(f"f'({x_point:.2f}) = {f_prime(x_point):.2f}", xy = (x_point, f(x_point)), xytext = (x_point - 1.5, f(x_point) - 1.5), arrowprops = dict(facecolor = 'black', arrowstyle = '->'), fontsize = 10, color = 'black')


    # Выделяем точку касания
    plt.scatter(x_point, f(x_point), color='black', zorder=5)

    # Выделяем оси
    plt.axvline(0, color='black', linewidth = 0.5)
    plt.axhline(0, color='black', linewidth = 0.5)

    # Добавляем подписи, заголовок и легенду
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title(f'Function, Derivative, and Tangent Line$')
    plt.legend(loc='upper right', prop={'size': 6})
    plt.grid(True)

    # Устанавливаем пределы на оси x и у
    plt.xlim(-2 * np.pi, 2 * np.pi)
    plt.ylim(-4, 4)

    # Показываем график
    plt.show()

# Пример использования
beauty_derivative_plot(3)
````

## Problem 2

The data in [populations.txt](https://drive.google.com/file/d/1Bq7qVzbZIOd058VobGE6iYjUcMFkbZMt/view?usp=drive_link) describes the populations of hares, lynxes and carrots in northern Canada during 20 years. Do the following exercises **without for-loops**.

- Plot the population size of each species for each year on a single graph. Add legend and axis labels to the plot.

````python
# Загрузка данных
data = np.loadtxt("populations.txt", skiprows=1)
years = data[:, 0]
hares = data[:, 1]
lynxes = data[:, 2]
carrots = data[:, 3]

# Построение графика
plt.plot(years, hares, label="Зайцы")
plt.plot(years, lynxes, label="Рыси")
plt.plot(years, carrots, label="Морковь")

# Добавление легенды и подписей
plt.xticks(np.arange(min(years), max(years) + 1, 2), rotation=45)
plt.xlabel("Год")
plt.ylabel("Численность популяции")
plt.legend()
plt.title("Популяции зайцев, рысей и моркови")
plt.grid(True)
plt.show()
````

- Find which year each species had the largest population. Mark these years in the plot of populations.

````python
# Находим годы с максимальной популяцией
max_hare_year = years[np.argmax(hares)]
max_lynx_year = years[np.argmax(lynxes)]
max_carrot_year = years[np.argmax(carrots)]

# Отмечаем точки на графике
plt.plot(years, hares, label="Зайцы")
plt.plot(years, lynxes, label="Рыси")
plt.plot(years, carrots, label="Морковь")
plt.xticks(np.arange(min(years), max(years) + 1, 2), rotation=45)
plt.scatter(max_hare_year, np.max(hares), color='blue', zorder=5)
plt.scatter(max_lynx_year, np.max(lynxes), color='orange', zorder=5)
plt.scatter(max_carrot_year, np.max(carrots), color='green', zorder=5)

# Опускаем перпендикуляры до оси Ox
plt.axvline(x=max_hare_year, color='blue', linestyle='--', alpha=0.5)  # Для зайцев
plt.axvline(x=max_lynx_year, color='orange', linestyle='--', alpha=0.5)  # Для рысей
plt.axvline(x=max_carrot_year, color='green', linestyle='--', alpha=0.5)  # Для моркови


# Добавление легенды и подписей
plt.annotate(f"Макс. зайцы: {int(np.max(hares))} в {int(max_hare_year)}", (max_hare_year, np.max(hares)), textcoords="offset points", xytext=(0,3), ha='left')
plt.annotate(f"Макс. рыси: {int(np.max(lynxes))} в {int(max_lynx_year)}", (max_lynx_year, np.max(lynxes)), textcoords="offset points", xytext=(0,3), ha='center')
plt.annotate(f"Макс. морковь: {int(np.max(carrots))} в {int(max_carrot_year)}", (max_carrot_year, np.max(carrots)), textcoords="offset points", xytext=(0,3), ha='left')
plt.xlabel("Год")
plt.ylabel("Численность популяции")
plt.legend()
plt.title("Популяции зайцев, рысей и моркови")
plt.grid(True)
plt.show()
plt.show()
````

- Assuming total population of hares and lynxes is 100%, plot stacked bar graph showing ratio of each specis for each year in the period.

````python
# Вычисляем общую популяцию зайцев и рысей
total_population = hares + lynxes
hare_ratio = hares / total_population * 100
lynx_ratio = lynxes / total_population * 100

# Построение stacked bar graph
plt.bar(years, hare_ratio, label="Зайцы")
plt.bar(years, lynx_ratio, bottom=hare_ratio, label="Рыси")

# Подписи годов с шагом 2
plt.xticks(np.arange(min(years), max(years) + 1, 2), rotation=45)
# Добавляем подписи и легенду
plt.xlabel("Год")
plt.ylabel("Доля популяции (%)")
plt.legend()
plt.title("Соотношение зайцев и рысей в общей популяции")
plt.grid(True)
plt.show()
````

- Find the mean and std of the populations of each species. Plot the historgam of population for each species and show mean values with vertical line. Show each histogram in a separate plot. Arrange 3 plots in a row.

````python
# Вычисляем среднее и стандартное отклонение
mean_hares = np.mean(hares)
std_hares = np.std(hares)
mean_lynxes = np.mean(lynxes)
std_lynxes = np.std(lynxes)
mean_carrots = np.mean(carrots)
std_carrots = np.std(carrots)

# Строим гистограммы
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.hist(hares, bins=10, color='blue', alpha=0.7)
plt.axvline(mean_hares, color='red', linestyle='--', label=f"Среднее: {mean_hares:.2f}")
plt.title("Зайцы")
plt.legend()

plt.subplot(1, 3, 2)
plt.hist(lynxes, bins=10, color='orange', alpha=0.7)
plt.axvline(mean_lynxes, color='red', linestyle='--', label=f"Среднее: {mean_lynxes:.2f}")
plt.title("Рыси")
plt.legend()

plt.subplot(1, 3, 3)
plt.hist(carrots, bins=10, color='green', alpha=0.7)
plt.axvline(mean_carrots, color='red', linestyle='--', label=f"Среднее: {mean_carrots:.2f}")
plt.title("Морковь")
plt.legend()

plt.tight_layout()
plt.show()
````

- Find which species (hares or lynxes) has the largest population for each year. Print the result as [H, H, L, H, ...]. Plot a pie chart showing the ratio of "H" and "L" values.

````python
# Ищем вид с наибольшей популяцией
dominant_species = np.where(hares > lynxes, "H", "L")
print(dominant_species)

# Считаем количество "H" и "L"
h_count = np.sum(dominant_species == "H")
l_count = np.sum(dominant_species == "L")

# Строим pie chart
plt.pie([h_count, l_count], labels=["Зайцы", "Рыси"], autopct='%1.1f%%')
plt.title("Соотношение лет с доминированием зайцев и рысей")
plt.show()
````

- Plot the change in the hare population and in the lynx population for each year. Find the correlation coefficient between the both time series.

````python
hare_changes = np.diff(hares)
lynx_changes = np.diff(lynxes)

plt.plot(years[1:], hare_changes, label="Изменение зайцев")
plt.plot(years[1:], lynx_changes, label="Изменение рысей")
plt.xlabel("Год")
plt.ylabel("Изменение популяции")
plt.legend()
plt.title("Изменение популяций зайцев и рысей")
plt.xticks(np.arange(min(years), max(years) + 1, 2), rotation=45)
plt.grid(True)
plt.show()

correlation = np.corrcoef(hare_changes, lynx_changes)[0, 1]
print(f"Коэффициент корреляции: {correlation:.2f}")
````

- Assume the population of hares in 1920 is unknown. Suggest a way to estimate this value. Compare an estimated value
with the true value and print a ratio of the error to the true value.

````python
from sklearn.linear_model import LinearRegression

# Обучаем модель на данных до 1920 года
X = years[:-1].reshape(-1, 1)
y = hares[:-1]
model = LinearRegression()
model.fit(X, y)

# Предсказываем популяцию за 1920 год
predicted_hare_1920 = model.predict([[1920]])[0]
true_hare_1920 = hares[-1]

# Вычисляем ошибку
error_ratio = abs(predicted_hare_1920 - true_hare_1920) / true_hare_1920
print(f"Предсказанное значение популяции зайцев в 1920 году: {predicted_hare_1920:.2f}")
print(f"Реальное значениепопуляции зайцев в 1920 году: {true_hare_1920}")
print(f"Ошибка: {error_ratio:.2%}")
````
